```python
import pandas as pd
import numpy as np
from pydataset import data
import seaborn as sns
```

```python
titanic = data('titanic')


# Tabla de contingencia class / survived
abs_tab = pd.crosstab(index=titanic['survived'],
            columns=titanic['class'], margins=True)
rel_tab = abs_tab.div(abs_tab['All']['All'])

print(abs_tab.div(abs_tab.loc['All']))
rel_tab[:-1].T.plot.bar()

relcum_tab = rel_tab.cumsum()
relcum_tab[:-1].T[:-1].plot.bar()
```
